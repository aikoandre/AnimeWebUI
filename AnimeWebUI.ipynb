```python
#@title üöÄ `Launch web UI`

#@markdown ### ‚≠ê Model selection <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Selecting-a-model)</sup>
#@markdown <small> It could take 4-7 minutes to see a link to the web UI, please be patient! :) </small>
model = "Animagine-XL 4.0 (SDXL)" # @param ["Animagine-XL 4.0 (SDXL)","Stable Diffusion 3.5 Large (SD3)","Stable Diffusion XL (SDXL)","FLUX.1-dev (FLUX)","Hassaku XL (SDXL)","WAI-NSFW-illustrious-SDXL (Illustrious)","Unholy Desire Mix (SDXL)","JANKU v4 NSFW Trained + NoobAI EPS + RouWei Illustrious XL (Illustrious)","Nova Anime XL (SDXL)"]

#@markdown ### üé® LoRA selection - Detailer <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LoRAs that enhance image quality and details </small>
detailer_lora_1 = "None" #@param ["None","TrendCraft (SDXL Base)","TrendCraft (Noob/Illustrious Base)"]
detailer_lora_2 = "None" #@param ["None","TrendCraft (SDXL Base)","TrendCraft (Noob/Illustrious Base)"]

#@markdown ### üé≠ LoRA selection - Style <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LoRAs that change the artistic style </small>
style_lora_1 = "None" #@param ["None","T-Rex Studio V2 NEW!!- Hentai +18","Flatline - Illus x Niji","Cartoony Anime Style","Bluethebone Style","Nyantcha / ThiccWithaQ Style"]
style_lora_2 = "None" #@param ["None","T-Rex Studio V2 NEW!!- Hentai +18","Flatline - Illus x Niji","Cartoony Anime Style","Bluethebone Style","Nyantcha / ThiccWithaQ Style"]

#@markdown ### üé¨ LoRA selection - Actions <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LoRAs for poses, actions, and specific movements </small>
action_lora_1 = "None" #@param ["None","DX888-Actions","Tucking Hair","Jack-O Challenge","Pole Dancing","All Fours","Playing with own hair","Smoking"]
action_lora_2 = "None" #@param ["None","DX888-Actions","Tucking Hair","Jack-O Challenge","Pole Dancing","All Fours","Playing with own hair","Smoking"]

#@markdown ### üõ†Ô∏è LoRA selection - Tools/Utility <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LoRAs for technical utilities and visual effects </small>
tool_lora_1 = "None" #@param ["None","Character Design Sheet","Neurocore Anime Shadow Circuit","Vivid Color Detail","Selfie","Underwater"]
tool_lora_2 = "None" #@param ["None","Character Design Sheet","Neurocore Anime Shadow Circuit","Vivid Color Detail","Selfie","Underwater"]

#@markdown ### üë• LoRA selection - Characters <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LoRAs for specific characters </small>
character_lora_1 = "Edelgard von Hresvelg" #@param ["None","Edelgard von Hresvelg","Shadowheart","Jonathan Joestar's"]
character_lora_2 = "None" #@param ["None","Edelgard von Hresvelg","Shadowheart","Jonathan Joestar's"]

#@markdown ### üëó LYCORIS selection - Clothing <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> LYCORIS models for clothing and outfits </small>
clothing_lycoris_1 = "None" #@param ["None","School Uniform","Track Uniform","Pajamas","Belly Dance Clothes","School Gym Uniform","Kunoichi","Yukata","Dress","Fantasy Mix Armor","Fantasy Mix Armor Alt","Fantasy Holy Armor","Fantasy Devil Armor","Formal Style","Furisode Hakama","Glow Dress","Modakawa Dress","Yomorio Lingerie","Bunny Girl","Oversized Shirt","Off Shoulder Sweater","Wafuku Gothic","Maid Dress"]
clothing_lycoris_2 = "None" #@param ["None","School Uniform","Track Uniform","Pajamas","Belly Dance Clothes","School Gym Uniform","Kunoichi","Yukata","Dress","Fantasy Mix Armor","Fantasy Mix Armor Alt","Fantasy Holy Armor","Fantasy Devil Armor","Formal Style","Furisode Hakama","Glow Dress","Modakawa Dress","Yomorio Lingerie","Bunny Girl","Oversized Shirt","Off Shoulder Sweater","Wafuku Gothic","Maid Dress"]

#@markdown ### üìñ Embedding selection <sup>[**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-LoRAs)</sup>
#@markdown <small> Embeddings for prompt enhancement </small>
embedding_1 = "None" #@param ["None","Lazy Embeddings Positive","Lazy Embeddings Negative"]
embedding_2 = "None" #@param ["None","Lazy Embeddings Positive","Lazy Embeddings Negative"]

#@markdown <small> Select ControlNet models to use [**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Selecting-ControlNet-models) </small>
controlnet = "v1.1" #@param ['none','v1.0','v1.0-diff','t2i','v1.1']
#@markdown ### üîß Web UI settings
#@markdown <small> This option only affects the **first** launch [**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/First-launch-web-UI-settings) </small>
webui_version = "latest" #@param ["stable","latest","latest-dev","ui-redesign"]
extensions_version = "latest" #@param ["none","lite","stable","latest","experimental"]
#@markdown <small> This saves your generation to Google Drive [**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Saving-outputs-to-Google-Drive) </small>
outputs_to_drive = False #@param {type:'boolean'}
output_drive_folder = "AI/Generated" #@param {type:'string'}
#@markdown <small> Change tunnels if you have connection issues with `gradio` [**[?]**](https://github.com/NUROISEA/anime-webui-colab/wiki/Using-different-tunnels) </small>
tunnel = "gradio" #@param ['gradio', 'ngrok', 'cloudflared', 'localhostrun', 'remotemoe']
ngrok_token = "" #@param { type:'string' }
ngrok_region = "auto" #@param ["auto", "ap", "au", "eu", "in", "jp", "sa", "us"]

# CivitAI API Configuration
CIVITAI_API_KEY = "f37aa076c4ebc0059ca9205f538e63ff"

# Extract model name without base info for internal use
model_name_clean = model.split(" (")[0]

# Model architecture mapping (clean names) - Enhanced with SD1.5 support
model_base_architecture = {
  "Animagine-XL 4.0": "SDXL",
  "Stable Diffusion 3.5 Large": "SD3",
  "Stable Diffusion XL": "SDXL",
  "FLUX.1-dev": "FLUX",
  "Hassaku XL": "SDXL",
  "WAI-NSFW-illustrious-SDXL": "Illustrious",
  "Unholy Desire Mix": "SDXL",
  "JANKU v4 NSFW Trained + NoobAI EPS + RouWei Illustrious XL": "Illustrious",
  "Nova Anime XL": "SDXL"
}

# Updated model dictionary (clean names)
model_dictionary = {
  # HuggingFace Models
  "Animagine-XL 4.0" : "https://huggingface.co/cagliostrolab/animagine-xl-4.0/resolve/main/animagine-xl-4.0.safetensors",
  "Stable Diffusion 3.5 Large" : "https://huggingface.co/stabilityai/stable-diffusion-3.5-large/resolve/main/sd3.5_large.safetensors",
  "Stable Diffusion XL" : "https://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/resolve/main/sd_xl_base_1.0.safetensors",
  "FLUX.1-dev" : "https://huggingface.co/black-forest-labs/FLUX.1-dev/resolve/main/flux1-dev.safetensors",

  # CivitAI Models (using model IDs)
  "Hassaku XL" : "575495",
  "WAI-NSFW-illustrious-SDXL" : "1367422",
  "Unholy Desire Mix" : "1723898",
  "JANKU v4 NSFW Trained + NoobAI EPS + RouWei Illustrious XL" : "1772645",
  "Nova Anime XL" : "1854228"
}

# Embeddings collection - Enhanced compatibility (most embeddings work across architectures)
embeddings = {
  "Lazy Embeddings Positive": {
    "id": "1833157",
    "compatible_bases": ["SD1.5", "SDXL", "Illustrious", "FLUX", "SD3"],  # Universal compatibility
    "description": "üìñ Positive prompt enhancement embedding - Works with all architectures",
    "category": "embedding",
    "priority": "high"  # High priority for quality improvement
  },
  "Lazy Embeddings Negative": {
    "id": "1860747", 
    "compatible_bases": ["SD1.5", "SDXL", "Illustrious", "FLUX", "SD3"],  # Universal compatibility
    "description": "üìñ Negative prompt enhancement embedding - Works with all architectures",
    "category": "embedding",
    "priority": "high"  # High priority for quality improvement
  }
}

# Reorganized LoRA collections with enhanced compatibility
detailer_loras = {
  "TrendCraft (SDXL Base)": {
    "id": "1768080",
    "compatible_bases": ["SDXL"],  # Specifically trained for SDXL
    "description": "üîß Enhances trending anime styles and details for SDXL architecture",
    "category": "detailer",
    "priority": "medium"
  },
  "TrendCraft (Noob/Illustrious Base)": {
    "id": "1809575",
    "compatible_bases": ["Illustrious"],  # Specifically for Illustrious/Noob models
    "description": "üîß Enhances trending anime styles and details for Illustrious architecture",
    "category": "detailer",
    "priority": "medium"
  }
}

style_loras = {
  "T-Rex Studio V2 NEW!!- Hentai +18": {
    "id": "1804885",
    "compatible_bases": ["Illustrious", "SDXL", "SD1.5"],  # Broader compatibility for style LoRAs
    "description": "üé® Adult/Hentai artistic style - Multi-architecture support",
    "category": "style",
    "priority": "low",
    "content_warning": "adult"
  },
  "Flatline - Illus x Niji": {
    "id": "1805073",
    "compatible_bases": ["Illustrious", "SDXL"],  # May work with SDXL too
    "description": "üé® Illustration x Niji style blend - Anime art style",
    "category": "style",
    "priority": "medium"
  },
  "Cartoony Anime Style": {
    "id": "1118776",
    "compatible_bases": ["Illustrious", "SDXL", "SD1.5"],  # General style, broader compatibility
    "description": "üé® Cartoon anime style - Universal anime aesthetic",
    "category": "style",
    "priority": "medium"
  },
  "Bluethebone Style": {
    "id": "1085590",
    "compatible_bases": ["Illustrious", "SDXL"],  # Artist style, may work with SDXL
    "description": "üé® Bluethebone artistic style - Distinctive art aesthetic",
    "category": "style",
    "priority": "medium"
  },
  "Nyantcha / ThiccWithaQ Style": {
    "id": "1192012",
    "compatible_bases": ["Illustrious", "SDXL"],  # Artist style, may work with SDXL
    "description": "üé® Nyantcha/ThiccWithaQ artistic style - Popular artist style",
    "category": "style",
    "priority": "medium"
  }
}

# Enhanced Actions category with improved compatibility
action_loras = {
  "DX888-Actions": {
    "id": "1875827",
    "compatible_bases": ["Illustrious"],  # Archive format, stay conservative
    "description": "üé¨ Enhanced action poses and dynamic movements",
    "category": "action",
    "note": "Archive format - contains multiple action pose LoRAs",
    "priority": "high"
  },
  "Tucking Hair": {
    "id": "1778841",
    "compatible_bases": ["Illustrious", "SDXL"],  # Pose LoRAs often work across architectures
    "description": "üé¨ Hair tucking behind ear pose and gesture",
    "category": "action",
    "priority": "medium"
  },
  "Jack-O Challenge": {
    "id": "1302043",
    "compatible_bases": ["Illustrious", "SDXL"],  # Specific pose, may work with SDXL
    "description": "üé¨ Specific dynamic pose from Guilty Gear character",
    "category": "action",
    "priority": "low"
  },
  "Pole Dancing": {
    "id": "1134393",
    "compatible_bases": ["Illustrious"],
    "description": "üé¨ Pole dancing poses and movements for Illustrious",
    "category": "action",
    "priority": "medium"
  },
  "Pole Dancing (SDXL)": {
    "id": "1394514",
    "compatible_bases": ["SDXL"],
    "description": "üé¨ Pole dancing poses and movements for SDXL",
    "category": "action",
    "priority": "medium"
  },
  "All Fours": {
    "id": "1708925",
    "compatible_bases": ["Illustrious", "SDXL"],  # Basic pose, likely broader compatibility
    "description": "üé¨ All fours crawling poses",
    "category": "action",
    "priority": "medium"
  },
  "Playing with own hair": {
    "id": "974347",
    "compatible_bases": ["Illustrious", "SDXL"],  # Gesture LoRA, may work with SDXL
    "description": "üé¨ Hair playing gestures and poses",
    "category": "action",
    "priority": "medium"
  },
  "Smoking": {
    "id": "1191354",
    "compatible_bases": ["Illustrious"],
    "description": "üé¨ Smoking poses and gestures for Illustrious",
    "category": "action",
    "priority": "low"
  },
  "Smoking (SDXL)": {
    "id": "953724",
    "compatible_bases": ["SDXL"],
    "description": "üé¨ Smoking poses and gestures for SDXL",
    "category": "action",
    "priority": "low"
  }
}

# Enhanced Tools/Utility category with improved compatibility
tool_loras = {
  "Character Design Sheet": {
    "id": "1096293",
    "compatible_bases": ["Illustrious", "SDXL"],  # Layout LoRA, may work with SDXL
    "description": "üõ†Ô∏è Character design sheet layout - Multi-view character sheets",
    "category": "tool",
    "priority": "high"
  },
  "Character Design Sheet (FLUX1.D)": {
    "id": "765872",
    "compatible_bases": ["FLUX"],
    "description": "üõ†Ô∏è Character design sheet layout for FLUX",
    "category": "tool",
    "priority": "high"
  },
  "Neurocore Anime Shadow Circuit": {
    "id": "1380520",
    "compatible_bases": ["Illustrious", "SDXL"],  # Visual effect, may work with SDXL
    "description": "üõ†Ô∏è Cyberpunk neural circuit effects",
    "category": "tool",
    "priority": "medium"
  },
  "Neurocore Anime Shadow Circuit (FLUX1.D)": {
    "id": "1050932",
    "compatible_bases": ["FLUX"],
    "description": "üõ†Ô∏è Cyberpunk neural circuit effects for FLUX",
    "category": "tool",
    "priority": "medium"
  },
  "Vivid Color Detail": {
    "id": "1657469",
    "compatible_bases": ["Illustrious", "SDXL", "SD1.5"],  # Quality enhancer, broader compatibility
    "description": "üõ†Ô∏è Enhanced color vibrancy and detail - Universal quality booster",
    "category": "tool",
    "priority": "high"
  },
  "Selfie": {
    "id": "972164",
    "compatible_bases": ["Illustrious", "SDXL"],  # Camera angle, may work with SDXL
    "description": "üõ†Ô∏è Selfie-style camera perspective",
    "category": "tool",
    "priority": "medium"
  },
  "Selfie (SDXL)": {
    "id": "1381634",
    "compatible_bases": ["SDXL"],
    "description": "üõ†Ô∏è Selfie-style camera perspective for SDXL",
    "category": "tool",
    "priority": "medium"
  },
  "Underwater": {
    "id": "1016010",
    "compatible_bases": ["Illustrious", "SDXL"],  # Environment effect, may work with SDXL
    "description": "üõ†Ô∏è Underwater scene effects and lighting",
    "category": "tool",
    "priority": "medium"
  },
  "Underwater (SDXL)": {
    "id": "818035",
    "compatible_bases": ["SDXL"],
    "description": "üõ†Ô∏è Underwater scene effects for SDXL",
    "category": "tool",
    "priority": "medium"
  }
}

character_loras = {
  "Edelgard von Hresvelg": {
    "id": "1305605",
    "compatible_bases": ["Illustrious", "SDXL"],  # Character LoRAs often work across architectures
    "description": "üë• Fire Emblem character - Edelgard von Hresvelg",
    "category": "character",
    "priority": "low"
  },
  "Shadowheart": {
    "id": "1542158",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üë• Baldur's Gate 3 character - Shadowheart",
    "category": "character",
    "priority": "low"
  },
  "Jonathan Joestar's": {
    "id": "1778951",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üë• JoJo's Bizarre Adventure character - Jonathan Joestar",
    "category": "character",
    "priority": "low"
  }
}

# Enhanced LYCORIS collection with improved compatibility matrix
clothing_lycoris = {
  "School Uniform": {
    "id": "1088202",
    "compatible_bases": ["Illustrious", "SDXL"],  # LYCORIS often more compatible
    "description": "üëó School uniform clothing",
    "category": "lycoris",
    "priority": "medium"
  },
  "School Uniform (SDXL)": {
    "id": "1555567",
    "compatible_bases": ["SDXL"],
    "description": "üëó School uniform clothing for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Track Uniform": {
    "id": "1155933",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Track and field uniform",
    "category": "lycoris",
    "priority": "medium"
  },
  "Track Uniform (SDXL)": {
    "id": "493832",
    "compatible_bases": ["SDXL"],
    "description": "üëó Track and field uniform for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Pajamas": {
    "id": "1180667",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Pajama clothing",
    "category": "lycoris",
    "priority": "medium"
  },
  "Pajamas (SDXL)": {
    "id": "558123",
    "compatible_bases": ["SDXL"],
    "description": "üëó Pajama clothing for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Belly Dance Clothes": {
    "id": "1083498",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Belly dancing outfit",
    "category": "lycoris",
    "priority": "medium"
  },
  "Belly Dance Clothes (SDXL)": {
    "id": "495468",
    "compatible_bases": ["SDXL"],
    "description": "üëó Belly dancing outfit for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "School Gym Uniform": {
    "id": "1179697",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó School gym/PE uniform",
    "category": "lycoris",
    "priority": "medium"
  },
  "School Gym Uniform (SDXL)": {
    "id": "516239",
    "compatible_bases": ["SDXL"],
    "description": "üëó School gym/PE uniform for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Kunoichi": {
    "id": "1023303",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Female ninja outfit",
    "category": "lycoris",
    "priority": "medium"
  },
  "Kunoichi (SDXL)": {
    "id": "486967",
    "compatible_bases": ["SDXL"],
    "description": "üëó Female ninja outfit for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Yukata": {
    "id": "1318687",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Traditional Japanese summer garment",
    "category": "lycoris",
    "priority": "medium"
  },
  "Yukata (SDXL)": {
    "id": "527505",
    "compatible_bases": ["SDXL"],
    "description": "üëó Traditional Japanese summer garment for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Dress": {
    "id": "1180806",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó General dress clothing",
    "category": "lycoris",
    "priority": "medium"
  },
  "Dress (SDXL)": {
    "id": "503998",
    "compatible_bases": ["SDXL"],
    "description": "üëó General dress clothing for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Fantasy Mix Armor": {
    "id": "1551197",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Mixed fantasy armor sets",
    "category": "lycoris",
    "priority": "medium"
  },
  "Fantasy Mix Armor (SDXL)": {
    "id": "1705463",
    "compatible_bases": ["SDXL"],
    "description": "üëó Mixed fantasy armor sets for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Fantasy Mix Armor Alt (SDXL)": {
    "id": "598975",
    "compatible_bases": ["SDXL"],
    "description": "üëó Alternative fantasy armor sets for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Fantasy Holy Armor (SDXL)": {
    "id": "596915",
    "compatible_bases": ["SDXL"],
    "description": "üëó Holy/paladin armor for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Fantasy Devil Armor (SDXL)": {
    "id": "596936",
    "compatible_bases": ["SDXL"],
    "description": "üëó Devil/demonic armor for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Formal Style": {
    "id": "965931",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Formal business attire",
    "category": "lycoris",
    "priority": "medium"
  },
  "Formal Style (SDXL)": {
    "id": "497782",
    "compatible_bases": ["SDXL"],
    "description": "üëó Formal business attire for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Furisode Hakama": {
    "id": "1166032",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Traditional Japanese formal wear",
    "category": "lycoris",
    "priority": "medium"
  },
  "Furisode Hakama (SDXL)": {
    "id": "527903",
    "compatible_bases": ["SDXL"],
    "description": "üëó Traditional Japanese formal wear for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Glow Dress": {
    "id": "1107836",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Glowing/luminescent dress",
    "category": "lycoris",
    "priority": "medium"
  },
  "Modakawa Dress": {
    "id": "1069463",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Modakawa-style dress",
    "category": "lycoris",
    "priority": "medium"
  },
  "Yomorio Lingerie": {
    "id": "1058894",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Yomorio-style lingerie",
    "category": "lycoris",
    "priority": "medium"
  },
  "Bunny Girl": {
    "id": "1548006",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Bunny girl outfit",
    "category": "lycoris",
    "priority": "medium"
  },
  "Bunny Girl (SDXL)": {
    "id": "587441",
    "compatible_bases": ["SDXL"],
    "description": "üëó Bunny girl outfit for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Oversized Shirt": {
    "id": "1314570",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Oversized shirt clothing",
    "category": "lycoris",
    "priority": "medium"
  },
  "Oversized Shirt (SDXL)": {
    "id": "889811",
    "compatible_bases": ["SDXL"],
    "description": "üëó Oversized shirt clothing for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Off Shoulder Sweater": {
    "id": "1003172",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Off-shoulder sweater",
    "category": "lycoris",
    "priority": "medium"
  },
  "Off Shoulder Sweater (SDXL)": {
    "id": "886431",
    "compatible_bases": ["SDXL"],
    "description": "üëó Off-shoulder sweater for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Wafuku Gothic": {
    "id": "975981",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Japanese gothic fashion",
    "category": "lycoris",
    "priority": "medium"
  },
  "Wafuku Gothic (SDXL)": {
    "id": "496941",
    "compatible_bases": ["SDXL"],
    "description": "üëó Japanese gothic fashion for SDXL",
    "category": "lycoris",
    "priority": "medium"
  },
  "Maid Dress": {
    "id": "1319831",
    "compatible_bases": ["Illustrious", "SDXL"],
    "description": "üëó Maid outfit",
    "category": "lycoris",
    "priority": "medium"
  },
  "Maid Dress (SDXL)": {
    "id": "527996",
    "compatible_bases": ["SDXL"],
    "description": "üëó Maid outfit for SDXL",
    "category": "lycoris",
    "priority": "medium"
  }
}

# Combine all LoRAs and LYCORIS for easy access
all_loras = {**detailer_loras, **style_loras, **action_loras, **tool_loras, **character_loras}
all_lycoris = {**clothing_lycoris}
all_embeddings = {**embeddings}

def get_compatible_items_by_category(selected_model, category_dict, include_priority=False):
    """Get list of items compatible with the selected model for a specific category"""
    model_base = model_base_architecture.get(selected_model, "Unknown")
    compatible_items = ["None"]
    
    # Sort by priority if available (high > medium > low)
    priority_order = {"high": 3, "medium": 2, "low": 1}
    
    items_list = []
    for item_name, item_info in category_dict.items():
        if model_base in item_info["compatible_bases"]:
            priority = item_info.get("priority", "medium")
            items_list.append((item_name, priority_order.get(priority, 2)))
    
    # Sort by priority (highest first) then alphabetically
    items_list.sort(key=lambda x: (-x[1], x[0]))
    
    if include_priority:
        # Return with priority indicators
        for item_name, priority_score in items_list:
            priority_name = {3: "HIGH", 2: "MED", 1: "LOW"}[priority_score]
            compatible_items.append(f"{item_name} [{priority_name}]")
    else:
        compatible_items.extend([item[0] for item in items_list])
    
    return compatible_items

def get_compatibility_summary(selected_model):
    """Get detailed compatibility summary with recommendations"""
    model_base = model_base_architecture.get(selected_model, "Unknown")
    
    summary = {
        "model": selected_model,
        "base": model_base,
        "categories": {},
        "recommendations": [],
        "total_compatible": 0
    }
    
    # Check each category
    categories = [
        ("detailer", detailer_loras, "üîß"),
        ("style", style_loras, "üé®"),
        ("action", action_loras, "üé¨"),
        ("tool", tool_loras, "üõ†Ô∏è"),
        ("character", character_loras, "üë•"),
        ("clothing", clothing_lycoris, "üëó"),
        ("embedding", embeddings, "üìñ")
    ]
    
    for cat_name, cat_dict, icon in categories:
        compatible = get_compatible_items_by_category(selected_model, cat_dict)[1:]  # Remove "None"
        high_priority = [item for item, info in cat_dict.items() 
                        if model_base in info["compatible_bases"] and info.get("priority") == "high"]
        
        summary["categories"][cat_name] = {
            "icon": icon,
            "count": len(compatible),
            "items": compatible,
            "high_priority": high_priority
        }
        summary["total_compatible"] += len(compatible)
    
    # Generate recommendations
    if summary["categories"]["embedding"]["high_priority"]:
        summary["recommendations"].append("üèÜ HIGHLY RECOMMENDED: Use both Lazy Embeddings for better quality")
    
    if summary["categories"]["tool"]["high_priority"]:
        summary["recommendations"].append("üèÜ RECOMMENDED: Character Design Sheet for reference sheets")
    
    if model_base == "SDXL":
        summary["recommendations"].append("üí° TIP: Use TrendCraft (SDXL Base) for enhanced quality")
        summary["recommendations"].append("üí° TIP: Many LYCORIS have SDXL-specific versions")
    elif model_base == "Illustrious":
        summary["recommendations"].append("üí° TIP: Illustrious has the most compatible items")
        summary["recommendations"].append("üí° TIP: Use TrendCraft (Noob/Illustrious Base) for quality")
    elif model_base == "FLUX":
        summary["recommendations"].append("üí° TIP: FLUX has fewer LoRAs but excellent embeddings")
        summary["recommendations"].append("üí° TIP: Focus on Character Design Sheet and Neurocore for FLUX")
    elif model_base == "SD3":
        summary["recommendations"].append("üí° TIP: SD3 works best with embeddings for quality enhancement")
    
    return summary

def display_compatibility_info(selected_model):
    """Display enhanced compatibility information with recommendations"""
    summary = get_compatibility_summary(selected_model)
    
    print(f"üéØ Selected Model: {summary['model']}")
    print(f"üèóÔ∏è  Model Base Architecture: {summary['base']}")
    print(f"üìÖ Current Date and Time (UTC): 2025-06-12 04:23:02")
    print(f"üë§ Current User's Login: aikoandre")
    print()
    
    # Show compatibility by category with priorities
    for cat_name, cat_data in summary["categories"].items():
        icon = cat_data["icon"]
        count = cat_data["count"]
        high_priority_count = len(cat_data["high_priority"])
        
        priority_text = f" ({high_priority_count} high priority)" if high_priority_count > 0 else ""
        print(f"{icon} Compatible {cat_name.title()}: {count} items{priority_text}")
        
        if cat_data["high_priority"]:
            print(f"  üèÜ High Priority: {', '.join(cat_data['high_priority'])}")
    
    print(f"\n‚úÖ Total Compatible Items: {summary['total_compatible']}")
    
    if summary['total_compatible'] == 0:
        print(f"‚ö†Ô∏è  No compatible items found for {summary['base']} architecture")
        print(f"üí° Consider switching to an Illustrious or SDXL model for more options")
    
    # Show recommendations
    if summary["recommendations"]:
        print("\nüìù RECOMMENDATIONS:")
        for rec in summary["recommendations"]:
            print(f"   {rec}")
    
    print("\nüìà COMPATIBILITY MATRIX:")
    print("   ‚Ä¢ üìñ Embeddings: Universal (work with ALL architectures)")
    print("   ‚Ä¢ üîß Detailer LoRAs: Architecture-specific training")
    print("   ‚Ä¢ üé® Style LoRAs: Often cross-compatible between similar architectures")
    print("   ‚Ä¢ üé¨ Action/üõ†Ô∏è Tool LoRAs: Mixed compatibility based on training")
    print("   ‚Ä¢ üëó LYCORIS: Generally more architecture-flexible than LoRAs")
    print("   ‚Ä¢ üë• Character LoRAs: Often work across SDXL-family architectures")
    
    print("\nüÜï WHAT'S NEW:")
    print("   ‚Ä¢ üèÜ Priority system: Focus on HIGH priority items first")
    print("   ‚Ä¢ üîÑ Improved compatibility: More cross-architecture support")
    print("   ‚Ä¢ üìñ Universal embeddings: Work with all model types")
    print("   ‚Ä¢ üìä Smart recommendations: Tailored to your model choice")
    
    # Return data for processing
    return tuple(get_compatible_items_by_category(selected_model, cat_dict) 
                for cat_dict in [detailer_loras, style_loras, action_loras, tool_loras, 
                               character_loras, clothing_lycoris, embeddings])

import os
import glob
import time

def validate_file(filepath, min_size_mb=30):
    """Validate if downloaded file is valid (lowered threshold for character LoRAs)"""
    if not os.path.exists(filepath):
        return False

    file_size = os.path.getsize(filepath)
    min_size_bytes = min_size_mb * 1024 * 1024

    if file_size < min_size_bytes:
        print(f"‚ùå File too small: {file_size} bytes (minimum: {min_size_bytes} bytes)")
        return False

    # For archive files, just check size (DX888-Actions is an archive)
    if filepath.endswith('.zip') or filepath.endswith('.rar') or filepath.endswith('.7z'):
        print(f"‚úÖ Archive file validated: {os.path.basename(filepath)}")
        return True

    # Check if it's a valid safetensors file
    try:
        with open(filepath, 'rb') as f:
            header = f.read(100)
            if b'{"' not in header and b"'{" not in header:
                print(f"‚ùå Invalid safetensors format")
                return False
    except Exception as e:
        print(f"‚ùå Error reading file: {e}")
        return False

    return True

def fix_downloaded_files(directory):
    """Fix files that have .1, .2, etc. extensions added by wget"""
    pattern = os.path.join(directory, "*.*.*")
    broken_files = glob.glob(pattern)

    for broken_file in broken_files:
        if broken_file.endswith(('.1', '.2', '.3', '.4', '.5')):
            # Remove the numeric extension
            fixed_name = broken_file.rsplit('.', 1)[0]

            try:
                if os.path.exists(fixed_name):
                    os.remove(fixed_name)  # Remove existing file if any

                os.rename(broken_file, fixed_name)
                print(f"üîß Fixed filename: {os.path.basename(broken_file)} ‚Üí {os.path.basename(fixed_name)}")
                return os.path.basename(fixed_name)
            except Exception as e:
                print(f"‚ùå Error fixing filename: {e}")

    return None

def find_downloaded_file(directory, pattern="*"):
    """Find the most recently downloaded file (safetensors or archive)"""
    # First try to fix any broken filenames
    fixed_file = fix_downloaded_files(directory)
    if fixed_file:
        filepath = os.path.join(directory, fixed_file)
        if validate_file(filepath, min_size_mb=10):  # Lower threshold for archives
            return fixed_file

    # Then search for normal files
    patterns = ["*.safetensors", "*.zip", "*.rar", "*.7z"]
    files = []

    for pat in patterns:
        pattern_path = os.path.join(directory, pat)
        files.extend(glob.glob(pattern_path))

    if not files:
        print(f"‚ùå No compatible files found in {directory}")
        return None

    # Get the most recent file
    latest_file = max(files, key=os.path.getmtime)
    filename = os.path.basename(latest_file)

    print(f"üìÅ Found file: {filename}")

    # Determine validation threshold based on file type
    min_size = 100 if "models" in directory.lower() else 10
    if validate_file(latest_file, min_size_mb=min_size):
        print(f"‚úÖ Valid file: {filename}")
        return filename
    else:
        print(f"‚ùå Invalid file: {filename}")
        return None

def download_from_civitai(model_id, directory, api_key=CIVITAI_API_KEY):
    """Download from CivitAI using API key method with better file detection"""
    url = f"https://civitai.com/api/download/models/{model_id}?token={api_key}"

    print(f"üì• Downloading from CivitAI (ID: {model_id})...")

    # Get files before download
    files_before = set()
    if os.path.exists(directory):
        files_before = set(os.listdir(directory))
    else:
        os.makedirs(directory, exist_ok=True)

    try:
        # Download using your specified method
        !wget -P "{directory}" -q --show-progress --content-disposition "{url}"

        # Wait for download to complete
        time.sleep(3)

        # Fix any broken filenames first
        fixed_file = fix_downloaded_files(directory)
        if fixed_file:
            print(f"‚úÖ Successfully downloaded and fixed: {fixed_file}")
            return fixed_file

        # Check for new files
        files_after = set(os.listdir(directory))
        new_files = files_after - files_before

        if new_files:
            # Find compatible files (safetensors or archives)
            compatible_files = [f for f in new_files if any(ext in f for ext in ['.safetensors', '.zip', '.rar', '.7z'])]

            if compatible_files:
                downloaded_file = compatible_files[0]
                filepath = os.path.join(directory, downloaded_file)

                # Use appropriate validation threshold
                min_size = 100 if "models" in directory.lower() else 10
                if validate_file(filepath, min_size_mb=min_size):
                    print(f"‚úÖ Successfully downloaded: {downloaded_file}")
                    return downloaded_file
                else:
                    print(f"‚ùå Downloaded file validation failed: {downloaded_file}")
                    return None
            else:
                print(f"‚ùå No compatible files found in downloaded files: {new_files}")
                return None
        else:
            # No new files detected, try to find latest file
            print("‚ö†Ô∏è  No new files detected, searching for latest compatible file...")
            return find_downloaded_file(directory)

    except Exception as e:
        print(f"‚ùå CivitAI download error: {e}")
        return None

def download_from_huggingface(url, filepath):
    """Download from HuggingFace"""
    print(f"üì• Downloading from HuggingFace...")

    # Create directory if it doesn't exist
    os.makedirs(os.path.dirname(filepath), exist_ok=True)

    try:
        !wget --timeout=60 --tries=3 -q --show-progress "{url}" -O "{filepath}"

        if validate_file(filepath, min_size_mb=100):
            print(f"‚úÖ Successfully downloaded: {os.path.basename(filepath)}")
            return True
        else:
            print(f"‚ùå HuggingFace download validation failed")
            if os.path.exists(filepath):
                os.remove(filepath)
            return False

    except Exception as e:
        print(f"‚ùå HuggingFace download error: {e}")
        return False

def safe_download_model(model_name, model_info, models_dir):
    """Download model with proper handling for both HuggingFace and CivitAI"""

    # Ensure models directory exists
    os.makedirs(models_dir, exist_ok=True)

    # Check if it's a CivitAI model (model_info is just an ID)
    if isinstance(model_info, str) and model_info.isdigit():
        # CivitAI model
        downloaded_filename = download_from_civitai(model_info, models_dir)
        return downloaded_filename
    else:
        # HuggingFace model
        filename = model_info.split("/")[-1]
        filepath = os.path.join(models_dir, filename)

        # Remove existing invalid file
        if os.path.exists(filepath) and not validate_file(filepath, min_size_mb=100):
            print(f"üóëÔ∏è Removing invalid file: {filepath}")
            os.remove(filepath)

        # Download if not exists
        if not os.path.exists(filepath):
            if download_from_huggingface(model_info, filepath):
                return filename
            else:
                return None
        else:
            print(f"‚úÖ Valid file already exists: {filename}")
            return filename

def safe_download_lora(lora_name, lora_info, lora_dir):
    """Download LoRA with proper handling"""

    # Ensure LoRA directory exists
    os.makedirs(lora_dir, exist_ok=True)

    # lora_info is now a dict with id and compatibility info
    lora_id = lora_info["id"]
    downloaded_filename = download_from_civitai(lora_id, lora_dir)
    return downloaded_filename

# Display compatibility info for selected model
print("üîç Analyzing model compatibility...")
detailer_compat, style_compat, action_compat, tool_compat, character_compat, clothing_compat, embedding_compat = display_compatibility_info(model_name_clean)

model_info = model_dictionary[model_name_clean]

# Get selected items and validate compatibility
selected_loras = []
selected_lycoris = []
selected_embeddings = []
model_base = model_base_architecture.get(model_name_clean, "Unknown")

print(f"\nüéÆ Processing selected items...")

# Process detailer LoRAs
for lora_selection in [detailer_lora_1, detailer_lora_2]:
    if lora_selection != "None" and lora_selection in detailer_loras:
        lora_info = detailer_loras[lora_selection]
        if model_base in lora_info["compatible_bases"]:
            selected_loras.append(lora_selection)
            print(f"‚úÖ üîß {lora_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lora_selection} - not compatible with {model_base}")

# Process style LoRAs
for lora_selection in [style_lora_1, style_lora_2]:
    if lora_selection != "None" and lora_selection in style_loras:
        lora_info = style_loras[lora_selection]
        if model_base in lora_info["compatible_bases"]:
            selected_loras.append(lora_selection)
            print(f"‚úÖ üé® {lora_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lora_selection} - not compatible with {model_base}")

# Process action LoRAs
for lora_selection in [action_lora_1, action_lora_2]:
    if lora_selection != "None" and lora_selection in action_loras:
        lora_info = action_loras[lora_selection]
        if model_base in lora_info["compatible_bases"]:
            selected_loras.append(lora_selection)
            print(f"‚úÖ üé¨ {lora_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lora_selection} - not compatible with {model_base}")

# Process tool LoRAs (NEW CATEGORY)
for lora_selection in [tool_lora_1, tool_lora_2]:
    if lora_selection != "None" and lora_selection in tool_loras:
        lora_info = tool_loras[lora_selection]
        if model_base in lora_info["compatible_bases"]:
            selected_loras.append(lora_selection)
            print(f"‚úÖ üõ†Ô∏è {lora_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lora_selection} - not compatible with {model_base}")

# Process character LoRAs
for lora_selection in [character_lora_1, character_lora_2]:
    if lora_selection != "None" and lora_selection in character_loras:
        lora_info = character_loras[lora_selection]
        if model_base in lora_info["compatible_bases"]:
            selected_loras.append(lora_selection)
            print(f"‚úÖ üë• {lora_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lora_selection} - not compatible with {model_base}")

# Process clothing LYCORIS (NEW CATEGORY)
for lycoris_selection in [clothing_lycoris_1, clothing_lycoris_2]:
    if lycoris_selection != "None" and lycoris_selection in clothing_lycoris:
        lycoris_info = clothing_lycoris[lycoris_selection]
        if model_base in lycoris_info["compatible_bases"]:
            selected_lycoris.append(lycoris_selection)
            print(f"‚úÖ üëó {lycoris_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {lycoris_selection} - not compatible with {model_base}")

# Process embeddings (NEW CATEGORY)
for embedding_selection in [embedding_1, embedding_2]:
    if embedding_selection != "None" and embedding_selection in embeddings:
        embedding_info = embeddings[embedding_selection]
        if model_base in embedding_info["compatible_bases"]:
            selected_embeddings.append(embedding_selection)
            print(f"‚úÖ üìñ {embedding_selection} is compatible with {model_name_clean} ({model_base})")
        else:
            print(f"‚ö†Ô∏è  Skipping {embedding_selection} - not compatible with {model_base}")

vae_link = ""
vae_name = vae_link.split("/")[-1]

'''
There are a quite few quality tags, which are:
1. Standard v3.0:
  masterpiece, best quality
2. Standard v3.1:
  masterpiece, best quality, very aesthetic, absurdres
3. Light v3.1:
  (masterpiece), best quality, very aesthetic, perfect face
4. Heavy v3.1:
  (masterpiece), (best quality), (ultra-detailed), very aesthetic, illustration, disheveled hair, perfect composition, moist skin, intricate details
'''

ui_config_dictionary = {
  "txt2img/Prompt/value": "(masterpiece, best quality, very aesthetic), absurdres, ",
  "txt2img/Negative prompt/value": "nsfw, lowres, (bad), text, error, fewer, extra, missing, worst quality, jpeg artifacts, low quality, watermark, unfinished, displeasing, oldest, early, chromatic aberration, signature, extra digits, artistic error, username, scan, [abstract]",
  "txt2img/Sampling steps/value": 28,

  # SDXL resolutions, for some reason colab can handle these now
  "txt2img/Width/value": 768,
  "txt2img/Height/value": 1344,
}

config_dictionary = {
  # literally does nothing to SDXL models
  "CLIP_stop_at_last_layers": 1,
}

git_flags = "-q"
wget_flags = "-q --show-progress"

try:
  utility
except NameError:
  !wget -q https://github.com/NUROISEA/anime-webui-colab/raw/main/configs/utility.py -P /content/
  import utility
  utility.log_usage("notebook-animagine-xl")

utility.mount_drive(outputs_to_drive)

if not utility.has_run:

  !{utility.install_webui(webui_version)}

  # Enhanced extensions list with CLIP Interrogator
  enhanced_extensions_list = []
  for extension_link in utility.extensions_list(extensions_version,webui_version,controlnet):
    enhanced_extensions_list.append(extension_link)

  # Add CLIP Interrogator extension for prompt generation from images
  clip_interrogator_url = "https://github.com/pharmapsychotic/clip-interrogator-ext.git"
  enhanced_extensions_list.append(clip_interrogator_url)

  print(f"üì¶ Installing {len(enhanced_extensions_list)} extensions (including CLIP Interrogator)...")
  for extension_link in enhanced_extensions_list:
    extension_name = extension_link.split('/')[-1].replace('.git', '')
    print(f'  ‚îî {extension_name}')
    !git clone {git_flags} {extension_link}

  # Add additional useful extensions for anime generation
  print("üé® Installing additional anime-focused extensions...")

  # Additional Background Removal extension
  !git clone {git_flags} https://github.com/AUTOMATIC1111/stable-diffusion-webui-rembg.git
  print("  ‚îî stable-diffusion-webui-rembg (Background Removal)")

  # Ultimate SD Upscale extension
  !git clone {git_flags} https://github.com/Coyote-A/ultimate-upscale-for-automatic1111.git
  print("  ‚îî ultimate-upscale-for-automatic1111 (Advanced Upscaling)")

  for config_file in utility.configs_list():
    !wget -q {config_file} -P {utility.web_ui_folder}/

  for embedding in utility.embeddings_list():
    !wget -q {embedding} -P {utility.embeddings_folder}/

  utility.dictionary_to_json(f"{utility.web_ui_folder}/ui-config.json", ui_config_dictionary)
  utility.dictionary_to_json(f"{utility.web_ui_folder}/config.json", config_dictionary)

  for patch in utility.patch_list():
    !{patch}
  %env LD_PRELOAD=/content/libtcmalloc_minimal.so.4

  utility.has_run = True

utility.output_to_gdrive(outputs_to_drive, output_drive_folder)

# Clean up any corrupted files first
print("üßπ Cleaning up any corrupted files...")
!find {utility.models_folder} -name "*.safetensors" -size -1000M -delete 2>/dev/null || true

# Download selected model
print(f"üì• Downloading {model_name_clean}...")
downloaded_model = safe_download_model(model_name_clean, model_info, utility.models_folder)

if not downloaded_model:
    print(f"‚ùå Download failed for {model_name_clean}")
    print(f"‚ö†Ô∏è  Please try selecting a different model or check your internet connection.")
    exit(1)

model_name = downloaded_model
print(f"‚úÖ Model ready: {model_name}")

!{utility.download_vae(vae_link)}

# Download compatible LoRAs
if selected_loras:
    print(f"üé® Downloading {len(selected_loras)} compatible LoRAs...")
    for lora_name in selected_loras:
        lora_info = all_loras[lora_name]

        category_icons = {"detailer": "üîß", "style": "üé®", "action": "üé¨", "tool": "üõ†Ô∏è", "character": "üë•"}
        category_icon = category_icons[lora_info["category"]]
        print(f"  ‚îî {category_icon} Downloading {lora_name}...")
        print(f"    üìã {lora_info['description']}")

        # Show special note for archive files
        if "note" in lora_info:
            print(f"    ‚ÑπÔ∏è  {lora_info['note']}")

        downloaded_lora = safe_download_lora(lora_name, lora_info, f"{utility.web_ui_folder}/models/Lora")

        if downloaded_lora:
            print(f"  ‚úÖ {lora_name} downloaded as: {downloaded_lora}")
        else:
            print(f"  ‚ùå Failed to download {lora_name}")

    print(f"üéâ LoRA download process completed!")
else:
    print("‚ÑπÔ∏è  No compatible LoRAs selected for download.")

# Download compatible LYCORIS
if selected_lycoris:
    print(f"üëó Downloading {len(selected_lycoris)} compatible LYCORIS...")
    for lycoris_name in selected_lycoris:
        lycoris_info = all_lycoris[lycoris_name]

        print(f"  ‚îî üëó Downloading {lycoris_name}...")
        print(f"    üìã {lycoris_info['description']}")

        downloaded_lycoris = safe_download_lora(lycoris_name, lycoris_info, f"{utility.web_ui_folder}/models/Lora")

        if downloaded_lycoris:
            print(f"  ‚úÖ {lycoris_name} downloaded as: {downloaded_lycoris}")
        else:
            print(f"  ‚ùå Failed to download {lycoris_name}")

    print(f"üéâ LYCORIS download process completed!")
else:
    print("‚ÑπÔ∏è  No compatible LYCORIS selected for download.")

# Download compatible embeddings
if selected_embeddings:
    print(f"üìñ Downloading {len(selected_embeddings)} compatible embeddings...")
    for embedding_name in selected_embeddings:
        embedding_info = all_embeddings[embedding_name]

        print(f"  ‚îî üìñ Downloading {embedding_name}...")
        print(f"    üìã {embedding_info['description']}")

        downloaded_embedding = safe_download_lora(embedding_name, embedding_info, f"{utility.web_ui_folder}/embeddings")

        if downloaded_embedding:
            print(f"  ‚úÖ {embedding_name} downloaded as: {downloaded_embedding}")
        else:
            print(f"  ‚ùå Failed to download {embedding_name}")

    print(f"üéâ Embedding download process completed!")
else:
    print("‚ÑπÔ∏è  No compatible embeddings selected for download.")

# Enhanced ControlNet info display
if controlnet != "none":
    print(f"\nüéÆ ControlNet Configuration: {controlnet}")
    print("üìã ControlNet Models that will be available:")

    controlnet_info = {
        "v1.0": [
            "‚Ä¢ canny - Edge detection for line art",
            "‚Ä¢ depth - Depth estimation for 3D structure",
            "‚Ä¢ hed - Holistically-nested edge detection",
            "‚Ä¢ mlsd - Line segment detection",
            "‚Ä¢ normal - Normal map generation",
            "‚Ä¢ openpose - Human pose detection",
            "‚Ä¢ scribble - Rough sketch/scribble control",
            "‚Ä¢ seg - Semantic segmentation"
        ],
        "v1.0-diff": [
            "‚Ä¢ Same as v1.0 but with different/improved weights",
            "‚Ä¢ Often better compatibility with newer models"
        ],
        "t2i": [
            "‚Ä¢ t2i-adapter models",
            "‚Ä¢ Lightweight alternative to full ControlNet",
            "‚Ä¢ Faster processing"
        ],
        "v1.1": [
            "‚Ä¢ All v1.0 models plus improvements:",
            "‚Ä¢ tile - Tiled/upscaling control",
            "‚Ä¢ inpaint - Inpainting control",
            "‚Ä¢ instruct_pix2pix - Instruction-based editing",
            "‚Ä¢ ip2p - Image-to-image with prompts",
            "‚Ä¢ shuffle - Content shuffling",
            "‚Ä¢ Better accuracy and stability"
        ]
    }

    if controlnet in controlnet_info:
        for item in controlnet_info[controlnet]:
            print(f"    {item}")

    print("\nüí° ControlNet Usage Tips:")
    print("    ‚Ä¢ Use 'canny' for precise line art control")
    print("    ‚Ä¢ Use 'openpose' for character pose control")
    print("    ‚Ä¢ Use 'depth' for maintaining 3D structure")
    print("    ‚Ä¢ Use 'tile' for upscaling with detail preservation")

for controlnet_model_link in utility.controlnet_list(controlnet,webui_version,extensions_version):
  !{utility.download_controlnet(controlnet_model_link)}

%cd {utility.web_ui_folder}

arg_list = utility.arguments(
  model=f"{utility.models_folder}/{model_name}",
  vae=f"{utility.vae_folder}/{vae_name}",
  tunnel=tunnel,
  ng_token=ngrok_token,
  ng_region=ngrok_region,
)

args = " ".join(arg_list)
total_items = len(selected_loras) + len(selected_lycoris) + len(selected_embeddings)
print(f"\nüöÄ Launching WebUI with enhanced features:")
print(f"   üì∏ CLIP Interrogator - Generate prompts from images")
print(f"   üéÆ ControlNet ({controlnet}) - Advanced image control")
print(f"   üñºÔ∏è  Background Removal - Clean image backgrounds")
print(f"   üìà Ultimate Upscale - High-quality image upscaling")
print(f"   üéØ Model: {model_name}")
print(f"   üé® LoRAs: {len(selected_loras)} compatible LoRAs loaded")
print(f"   üëó LYCORIS: {len(selected_lycoris)} clothing models loaded")
print(f"   üìñ Embeddings: {len(selected_embeddings)} prompt enhancers loaded")
print(f"   üì¶ Total Items: {total_items} additional models ready")
print(f"\nüÜï NEW Categories Available:")
print(f"   ‚Ä¢ üõ†Ô∏è Tools/Utility - Character sheets, effects, camera angles")
print(f"   ‚Ä¢ üëó LYCORIS Clothing - Comprehensive outfit collection")
print(f"   ‚Ä¢ üìñ Embeddings - Lazy prompt enhancement for quality")
print(f"   ‚Ä¢ üé¨ Enhanced Actions - Pole dancing, poses, gestures")
print(f"   ‚Ä¢ üé® Expanded Styles - Bluethebone, Nyantcha styles")
!python launch.py {args}
```
